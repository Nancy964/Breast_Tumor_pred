# ENCODING THE TARGET VARIABLE


```python
# from sklearn.preprocessing import LabelEncoder
# Le=LabelEncoder()
```

```python
breast['diagnosis']=breast['diagnosis'].map({"M":1,"B":0})
```

```python
breast['diagnosis'].value_counts()
```

# SPLITTING DATA

```python
x=breast[['radius_mean','texture_mean',	'perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','symmetry_mean']]
y=breast['diagnosis']
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

# Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
```

```python
sc.fit(x_train)
```

```python
x_train=sc.transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
Lg=LogisticRegression()
```

```python
Lg.fit(x_train,y_train)
```

```python
y_pred=Lg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
